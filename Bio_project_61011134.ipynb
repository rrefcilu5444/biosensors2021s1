```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style="white") #white background style for seaborn plots
sns.set(style="whitegrid", color_codes=True)
```

```python
df = pd.read_csv('data_bio_present.csv')
df2 = pd.read_csv('data_bio_present_2.csv')
df3 = pd.read_csv('data_bio_present_3.csv')
df4 = pd.read_csv('data_bio_present_4.csv')
df.head()
```

```python
df.dtypes
```

```python
x = df['Day']
y = df['Pos']
plt.plot(x, y, color = 'r', linestyle = '--',
         marker = '.',label = "Positive")
plt.plot(x, df['Total'], color = 'g', linestyle = '--',
         marker = '.',label = "Total")
plt.plot(x, df['ATK'], color = 'c', linestyle = '-',
         label = "ATK")
plt.xlabel('day')
plt.ylabel('People')
plt.title('COVID-19 cases update daily')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x, df2['% Detech'], color = 'r', linestyle = '--',
         marker = '.',label = "% Detech")
plt.xlabel('Day')
plt.ylabel('% Detech')
plt.title('% Detech by RT-PCR per day')
plt.grid()
plt.legend()
plt.show()
```

```python
df3 = pd.read_csv('data_bio_present_3.csv')
x = df3['Day']
plt.plot(x, df3['ATK'], color = 'r', linestyle = '--',
         label = "ATK")
plt.xlabel('day')
plt.ylabel('Pos')
plt.title('Covid Detection in Thailand')
plt.grid()
plt.legend()
plt.show()
```

```python
x1 = df3['Day']
y1 = df3['ATK']
coeff = np.polyfit(x1,y1,1)
print(coeff)
```

```python
m = coeff[0]
b = coeff[1]
Trend = m*x1 + b
plt.plot(x1,Trend,'c-')
plt.xlabel('day')
plt.ylabel('Positive')
plt.title('Trend Line')
```

```python
df3 = pd.read_csv('data_bio_present_3.csv')
x = df3['Day']
plt.plot(x, df3['ATK'], color = 'r', linestyle = '--',
         label = "ATK")
plt.plot(x1,Trend,color = 'c', linestyle = '-',label = "Trend")
plt.xlabel('day')
plt.ylabel('Pos')
plt.title('ATK/Day')
plt.grid()
plt.legend()
plt.show()
```
